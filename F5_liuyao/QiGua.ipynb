```python


import random
import numpy as np

name2yao={u'老阴--X--':'0', u'少阳-----':'1',u'少阴-- --':'2',u'老阳--O--':'3'}
yao2name={'0':u'老阴--X--', '1':u'少阳-----', '2': u'少阴-- --', '3':u'老阳--O--'}
yao2bengua={'0':2, '1':1, '2':2, '3':1} # yao 
yao2biangua={'0':1, '1':1, '2':2, '3':2} #
yao2bian={'0':1, '1':0, '2':0, '3':1}
bengua2cuogua = {'1':2, '2':1, } # yao 

class Yao6:
    def __init__(self, gua):
        self.gua = list(gua) # input gua as a 6 character string
        self.gua_raw = [] # 六爻原始输入
        self.gua_name = [] # 六爻对应爻名
        self.gua_bian = [] # 六爻对应动静
        self.ben_gua = '' # 本卦：代表的是目前的情况。
        self.bian_gua = '' # 变卦：事情发展的结果。是由本卦经过某个爻的变动、变化而得到的，也就是变卦是由本卦变动而来的。
        self.hu_gua = '' # 互卦：事情发展的过程。本卦中的第三、四、五爻，拿出来作为互卦的上卦，本卦的第二、三、四爻，拿出来作为互卦的下卦。
        self.cuo_gua = '' # 错卦：一种从反面的角度。是与本卦阴阳全颠倒的卦，即每个爻 阴变阳、阳变阴而得到的卦。
        self.zong_gua = '' # 综卦：从另外一种角度来思考要测的事情。就是把本卦倒过来看。它们通常是从事物的反面来看事物的发展变化。
        self.bian_yao = [] # 变爻位置
        self.bu_bian_yao = [] # 不变爻位置
 
    def suan_gua(self):
        
        for i in range(6):
            self.gua_name.append(yao2name[self.gua[i]])
            self.gua_bian.append(yao2bian[self.gua[i]])
            self.gua_raw += str(self.gua[i])
            self.ben_gua += str(yao2bengua[self.gua[i]]) # 本卦，又叫主卦、原卦。它体现的是事物初始、开始阶段的 信息，或目前的情况。代表的是目前的情况
            self.bian_gua += str(yao2biangua[self.gua[i]]) # 变卦代表着事物发展变化的最终结果，也就是求测者想要得到的最终结果、结局。在预测时，事物的最终吉凶结果，要由变卦来作最后定夺。所以，预测最终要以变的结果为准。变卦为事情发展的结果
            if int(self.gua[i]) == 0 or int(self.gua[i]) == 3:
                self.bian_yao.append(i+1)
            else:
                self.bu_bian_yao.append(i+1)
        
        self.hu_gua = ''.join(self.ben_gua[1:4] + self.ben_gua[2:5])
        self.cuo_gua = ''.join([str(bengua2cuogua[y]) for y in self.ben_gua])
        self.zong_gua = ''.join(self.ben_gua[::-1])
        self.jie_gua_fa = jiegua[sum(self.gua_bian)]

    def jie_gua(self):
        print('===解卦===')
        print('本卦：' + gua64[self.ben_gua][0])
        if len(self.bian_yao) > 0:
            print('变卦：' + gua64[self.bian_gua][0])
        else:
            print('变卦：' + '无')

        zhu_gua = ['主要卦辞','辅助卦辞']
        zhu_yao = ['主要爻辞','辅助爻辞']
        gua_vec = [gua_sample.ben_gua, gua_sample.bian_gua]
        yao_vec = [self.bian_yao, self.bu_bian_yao]

        use_gua_ci = [None, None] # {0:None, 1:'zhu_gua', 2:'bian_gua'}
        use_yao_ci = [None, None] # ['zhu_yao', 'bian_yao']
        use_yong_jiu = None

        if len(self.bian_yao) == 0:
            use_gua_ci = [gua64[self.ben_gua], None]
        elif len(self.bian_yao) == 1:
            use_yao_ci = [gua64[self.ben_gua][self.bian_yao[0]], None]
        elif len(self.bian_yao) == 2:
            use_yao_ci = [gua64[self.ben_gua][idx] for idx in self.bian_yao[::-1]]
        elif len(self.bian_yao) == 3:
            use_gua_ci = [gua64[self.ben_gua], gua64[self.bian_gua]]
        elif len(self.bian_yao) == 4:
            use_yao_ci = [gua64[self.bian_gua][idx] for idx in self.bu_bian_yao]
        elif len(self.bian_yao) == 5:
            use_yao_ci = [gua64[self.bian_gua][self.bu_bian_yao[0]], None]
        elif len(self.bian_yao) == 6:
            if ''.join(self.gua) == '000000' or ''.join(self.gua) == '333333':
                use_yong_jiu = gua64[self.ben_gua][7]
            else:
                use_gua_ci = [gua64[self.ben_gua], None]
        else:
            print('TBA')

        print('')
        print('主要卦辞： ' + use_gua_ci[0][8]) if use_gua_ci[0] != None else print('主要卦辞： 无') 
        print('辅助卦辞： ' + use_gua_ci[1][8]) if use_gua_ci[1] != None else print('辅助卦辞： 无') 
        print('主要爻辞： ' + use_yao_ci[0]) if use_yao_ci[0] != None else print('主要爻辞： 无') 
        print('参考爻辞： ' + use_yao_ci[1]) if use_yao_ci[1] != None else print('参考爻辞： 无') 
        print('用九： ' + use_yong_jiu) if use_yong_jiu != None else print('') 


        print('本卦卦辞(目前的情况)： ' + gua64[self.ben_gua][8]) if self.ben_gua != None else print('主要卦辞： 无') 
        print('变卦卦辞(事情发展的结果)： ' + gua64[self.bian_gua][8]) if self.bian_gua != None else print('主要卦辞： 无') 
        print('互卦卦辞(事情发展的过程)： ' + gua64[self.hu_gua][8]) if self.hu_gua != None else print('主要卦辞： 无') 
        print('错卦卦辞(一种从反面的角度)： ' + gua64[self.cuo_gua][8]) if self.cuo_gua != None else print('辅助卦辞： 无') 
        print('综卦卦辞(另外一种角度来思考要测的事情)： ' + gua64[self.zong_gua][8]) if self.zong_gua != None else print('辅助卦辞： 无') 
        print('========')
        return self.jie_gua_fa
            

    def print(self):
        for i in reversed(range(len(self.gua_name))):
            print(self.gua_name[i])
```
